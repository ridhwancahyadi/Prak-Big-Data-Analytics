## Latihan


```python
df_brain = pd.read_csv('/content/drive/MyDrive/Prak Big Data Analytics/Per-12/brain_size.csv', delimiter=';',na_values='.')
df_brain = df_brain.iloc[:, 1:]
df_brain.head()
```

```python
df_brain.info()
```

### Hubungan Antar Variabel

```python
# isi data kosong
df_brain.isnull().sum()
```

```python
values = {'Height':f"{df_brain['Height'].mean():.1f}", 'Weight': f"{df_brain['Height'].mean():.1f}"}
df_new = df_brain.fillna(value=values)
df_new.head()
```

```python
sns.scatterplot(x='Height', y='Weight', hue='Gender', data=df_new)
```

```python
sns.pairplot(df_new, vars=['FSIQ', 'VIQ', 'PIQ'], hue='Gender')
```

```python
sns.lmplot(x='FSIQ', y='VIQ', hue='Gender', data=df_new)
```

```python
numerical_data = df_new.select_dtypes(include=['number'])
corr = numerical_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
