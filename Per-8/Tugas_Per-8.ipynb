```python
def stdev(nums):
  diffs = 0
  avg = sum(nums)/len(nums)
  for n in nums:
      diffs += (n - avg)**(2)
  return (diffs/(len(nums)-1))**(0.5)
```

```python
gaji_std = df_pegawai['gaji']
print(stdev(gaji_std))
```
