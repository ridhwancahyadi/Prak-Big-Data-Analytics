## Menghitung Ukuran Persebaran Data

```python
data = pd.read_csv('/content/drive/MyDrive/Prak Big Data Analytics/Per-8/brain_size.csv', sep=';', na_values='.')
```

```python
df = data
df
```

```python
def stdev(nums):
  diffs = 0
  avg = sum(nums)/len(nums)
  for n in nums:
      diffs += (n - avg)**(2)
  return (diffs/(len(nums)-1))**(0.5)
```

```python
viq = df['VIQ']
viq
```

```python
print(stdev(viq))
```

```python
viq.std()
```

```python
data_height = df['Height']
data_height
print(stdev(data_height))
```
